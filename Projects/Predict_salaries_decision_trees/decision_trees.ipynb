Predicting Salaries with Decision Trees
=======================================


![title](img/tree_5000_blank.png)

Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
attribute_names = (
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'salary-class',
)
```

Read a CSV data file.

```python
df_data = pd.read_csv('adult.data', names=attribute_names,sep=', ',engine='python')
df_data.head()
```

```python
df_data.info()
```

```python
mask = df_data.isnull()
```

```python
mask = mask.any(axis=1)
```

```python
df_data[mask]
```

Pre-process the data.  First, remove unwanted variables.

```python
excludes = [
    'fnlwgt', 
    'education-num',
    'capital-gain', 
    'capital-loss'
    ]

df_data.drop(columns= excludes,inplace=True)
```

```python
df_data.replace({'?': np.nan},inplace=True)
```

```python
df_data.dropna(inplace=True)
```

```python
maskd = df_data.duplicated(keep=False)
df_data[maskd]
```

```python
df_data['workclass'].value_counts()
```

Boolean variables to 0s and 1s.

```python
booleans = [
    'sex', 
    'native-country', 
    'salary-class'
]

for column in booleans:
    df_data[column]=df_data[column].astype('category').cat.codes
df_data.head()
```

 One-hot encoding of all categorical attributes.

```python
categoricals = [
    'workclass', 
    'education', 
    'marital-status', 
    'occupation', 
    'relationship', 
    'race'
]

dummie_data = pd.get_dummies(df_data[categoricals])
```

```python
data = pd.concat([df_data,dummie_data],axis=1)
```

```python
data.head()
```

```python
data.drop(columns=categoricals,inplace=True)
```

```python
data.head()
```

Extract attributes (X) and class labels (y).

```python
y=data['salary-class']
```

```python
X = data.copy()
x = X.drop(columns=['salary-class'])
```

Split dataset into training and testing subsets.

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=444)
print('Training shape: ',x_train.shape,y_train.shape)
print('Testing shape: ',x_test.shape,y_test.shape)
```

Construct a decision tree from this training data.

```python
dt = DecisionTreeClassifier(criterion='entropy',random_state=0).fit(x_train,y_train)
dt
```

Predict target class for the training set.

```python
predict_train = dt.predict(x_train)
print(predict_train)
train_acc = accuracy_score(y_train,predict_train)
print('Training accuracy: ', train_acc)
```

Predict target class for the testing set.

```python
predict_test = dt.predict(x_test)
print(predict_test)
test_acc = accuracy_score(y_test,predict_test)
print('Testing accuracy: ', test_acc)
```

Check if dataset is balanced.

```python
data[['salary-class']].value_counts()
```

Prune the decision tree.

```python
sizes = [400,500,750,900,1000,1500]
train_acc=[]
test_acc=[]

for s in sizes:
    clf = DecisionTreeClassifier(random_state=1,
                            criterion='entropy',
                                min_samples_split=s)
    clf.fit(x_train,y_train)
    train_pred = clf.predict(x_train)
    test_pred = clf.predict(x_test)
    train_acc.append(accuracy_score(y_train,train_pred))
    test_acc.append(accuracy_score(y_test,test_pred))
```

```python
df = pd.DataFrame(data={'size':sizes,'train_acc':train_acc,'test_acc':test_acc})
df
```

```python
plt.plot(sizes,train_acc,label='Train accuracy')
plt.plot(sizes,test_acc, label='Test accuracy')
plt.xlabel('Split size')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

Graphical representation of a decision tree.

```python
dt_clf = DecisionTreeClassifier(criterion='entropy',random_state=222,min_samples_split=900)
dt_clf.fit(x_train,y_train)
```

```python
train_pr = dt_clf.predict(x_train)
train_ac = accuracy_score(y_train,train_pr)
print('Training accuracy: ', train_ac)
test_pr = dt_clf.predict(x_test)
test_ac = accuracy_score(y_test,test_pr)
print('Validation accuracy: ', test_ac)
```

```python
fig = plt.figure(figsize=(40,40))
p = tree.plot_tree(dt_clf,
                  filled=True,
                  feature_names=x_train.columns.to_list(),
                  class_names=('LOW','HIGH'))
fig.savefig('tree.png')
```
