```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
X = torch.arange(-3,3,0.1).view(-1,1)
```

```python
f = X - 1
```

```python
Y = f + 0.1*torch.randn(X.size())
```

```python
plt.plot(X.numpy(),f.numpy())
plt.plot(X.numpy(),Y.numpy(),'xr')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
def cost_function(X,W,b):
    yhat = W*X+b
    return yhat
```

```python
def loss_function(Y,Yhat):
    loss = torch.mean((Yhat-Y)**2)
    return loss
```

```python
#Initializing the parameters W,b
W = torch.tensor(-15.,requires_grad=True)
b = torch.tensor(-10., requires_grad=True)
```

```python
def plot_fitted_line(Yhat,Y):
    
    plt.plot(Yhat.detach().numpy(),label='Predicted')
    plt.plot(Y.numpy(),'xr',label='True')
    plt.legend()
    plt.show()
```

```python
def plot_contour():
    w_range=np.linspace(-15,15,30)
    b_range=np.linspace(-15,15,30)
    w,b = np.meshgrid(w_range,b_range)
    
    Z = torch.mean((Y-w*X+b))
```

```python
LOSS_BGD = []
LOSS_SGD = []
def train_model(epochs,lr):
    for epoch in range(epochs):
        Yhat = cost_function(X,W,b)
        loss = loss_function(Y,Yhat)
        LOSS_BGD.append(loss.item())
        
        plot_fitted_line(Yhat,Y)
        
        loss.backward()
        
        W.data = W.data - lr*W.grad.data
        b.data = b.data - lr*b.grad.data
        
        W.grad.data.zero_()
        b.grad.data.zero_()
    
```
