```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = torch.tensor(2.0,requires_grad=True)
```

```python
y = X**2
```

```python
y.backward()
```

```python
X.grad
```

```python
# Partial derivatives
```

```python
u = torch.tensor(4.,requires_grad=True)
v = torch.tensor(2.,requires_grad=True)
```

```python
f = 2*u**2+v**3
```

```python
f.backward()
```

```python
u.grad
```

```python
v.grad
```

```python
X = torch.linspace(-np.pi,np.pi,100, requires_grad=True)
Y = torch.sin(X)
```

```python
y = Y.sum()
```

```python
y.backward()
```

```python
X.grad
```

```python
plt.plot(X.detach().numpy(),Y.detach().numpy())
plt.plot(X.detach().numpy(),X.grad.detach().numpy())
plt.show()
```

```python
Z = torch.linspace(-10,10,100,requires_grad=True)
A = torch.relu(Z)
a = A.sum()
```

```python
a.backward()
```

```python
Z.grad
```

```python
plt.plot(Z.detach().numpy(),A.detach().numpy())
plt.plot(Z.detach().numpy(),Z.grad.detach().numpy())
plt.show()
```
